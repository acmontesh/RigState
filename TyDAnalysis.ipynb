```python
import logging

class LoggerDev:

    def __init__( self,verbosity=2 ):
      for handler in logging.root.handlers[:]:
        logging.root.removeHandler(handler)
      logging.basicConfig(   format='[%(levelname)s]......... %(message)s', level=logging.DEBUG   )
      self.verbosity      =   verbosity

    def setVerbosity( self,verbosity ):
        self.verbosity      = verbosity
        logging.info( f"Verbosity has been set to {verbosity}" )

    def dbgMsg( self,msg=None ):
        if (msg is not None) and (self.verbosity>1):
            logging.debug(  msg  )

    def infoMsg( self,msg=None ):
        if (msg is not None) and (self.verbosity>1):
            logging.info(  msg  )

    def warningMsg( self,msg=None ):
        if (msg is not None) and (self.verbosity>0):
            logging.warning(  msg  )

    def errorMsg( self,msg=None ):
        if (msg is not None) and (self.verbosity>-1):
            logging.error(  msg  )

    def resultMessage( self,msg=None ):
        if (msg is not None) and (self.verbosity>1):
            logging.info(  f"[RESULTS] -------->  {msg}"  )

```

```python
from juliacall import Main as jl
jl.seval('include("./TyD_Model/TorqueAndDrag.jl")')

def runTaDSimulationCS( paths, rangeFrictionFactors=(0.01, 0.76, 0.05), stepResult=200 ):
    jl.seval("using .TnD")
    RIH,POOH,runDepths    =   jl.TnD.torqueAndDrag( paths["bha"],paths["fluid"],paths["wellbore"],paths["survey"],
                                                    rangeFrictionFactors[0],rangeFrictionFactors[1],rangeFrictionFactors[2],stepResult )
    return RIH,POOH,runDepths,rangeFrictionFactors[0],rangeFrictionFactors[1],rangeFrictionFactors[2]

def plotBroomstick( blockWeight,RIH,POOH,HKL_md,f1,f2,deltaF,featureDFs,customXLowerLim=0,customXUpperLim=1500,customDepthStep=1000, 
                        customXStep=200,saveFig=False,outFilePath="torque_and_drag_sim.png"   ):
    import matplotlib.pyplot as plt
    from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
    import numpy as np
    import pandas as pd
    from matplotlib.lines import Line2D
    from matplotlib.patches import Circle
    depthArrayFt    =   np.array(HKL_md)
    fig, ax         = plt.subplots(figsize=(3.33,4.6))
    for j,ff in np.ndenumerate( np.arange( f1,f2,deltaF ) ):
        HKLBuffer   = POOH[ :,j ]
        ax.plot( HKLBuffer/1000 + blockWeight,depthArrayFt, color='cyan',alpha=0.4 )
        HKLBuffer   = RIH[ :,j ]
        ax.plot( HKLBuffer/1000 + blockWeight,depthArrayFt, color='magenta',alpha=0.4 )
    if len( featureDFs )!=0:
        for df in featureDFs:
            ax.scatter( df["Hook Load [klb]"],df["Bit Depth [ft]"],facecolors="None",edgecolor='gold',s=20,zorder=2,marker="o",lw=0.6 )
    ax.set_ylim( 0.0,np.round(depthArrayFt.max(  )+customDepthStep) )
    ax.invert_yaxis( )
    ax.set_xlim( customXLowerLim,customXUpperLim )
    ax.xaxis.tick_top(  )
    ax.set_yticks(  np.arange(0.0,np.round(depthArrayFt.max(  )+customDepthStep),customDepthStep)  )
    ax.set_xticks(  np.arange(customXLowerLim,customXUpperLim+customXStep,customXStep)  )
    ax.xaxis.set_label_position('top')
    ax.set_xlabel(  "Hook Load [klb]", fontdict={'family':'arial', 'size':8, 'weight':'bold'}  )
    ax.set_ylabel(  "Measured Depth [ft]", fontdict={'family':'arial', 'size':8, 'weight':'bold'}  )
    ax.set_xticklabels(  ax.get_xticks(), fontdict={'family':'arial', 'size':8}  )
    ax.set_yticklabels(  ax.get_yticks(), fontdict={'family':'arial', 'size':8}  )          
    ax.grid(  color='darkgrey', linestyle='dotted', linewidth=1.0, which='major'  )
    ax.tick_params( 'x',rotation=90 )
    ax.xaxis.set_minor_locator( AutoMinorLocator(5) )
    costomLines     = [
                            Line2D(  [0], [0], color='cyan', lw=1.5  ),
                            Line2D(  [0], [0], color='magenta', lw=1.5  ),
                            Line2D(  [0], [0], marker='o', color='gold', markerfacecolor='none', markersize=10, linestyle=''  )
                        ]
    customLabels = [ 'Pick-Up Weight', 'Slack-Off Weight', 'Observed Hook Load' ]
    ax.legend( costomLines, customLabels, prop={'family':'arial', 'size':8},loc='upper right',facecolor='dimgray',labelcolor='white' )
    ax.set_facecolor('black')
    plt.tight_layout( pad=0.1 )
    if saveFig:  fig.savefig( outFilePath,dpi=600 )
```

```python
def getVariabilityBackCalcFrictionFactor( df,RIHA,POOHA,depths,f1,f2,deltaF,blockWeight ):
    import numpy as np
    import pandas as pd
    maxDiffHkl                                                  = 0
    maxDiffs                                                    = [ ]
    ffList                                                      = list( np.arange( f1,f2,deltaF ) )
    buffer                                                      = {  ff:{ } for ff in ffList  }
    xp                                                          = np.array( depths )
    bitDepthsArray,hookLoadArray                                =(df["Bit Depth [ft]"].values,df["Hook Load [klb]"].values)
    temporalDF                                                  = pd.DataFrame( {"Bit Depth [ft]": bitDepthsArray,
                                                                                  "Hook Load [klb]": hookLoadArray} )
    hookLoadArray                                               = hookLoadArray * 1000.
    bufferArray                                                 = np.array([ ])
    RIH                                                         = np.array( RIHA )
    POOH                                                        = np.array( POOHA )
    minHkls,maxHkls                                             = ([],[])
    for j,ff in enumerate( ffList ): 
        buffer[ff]['expected']                                  = np.interp( bitDepthsArray,xp,
                                                                                    POOH[ :,j ]  + blockWeight*1000 ) 
        if j==0:                                                minHkls = buffer[ff]['expected']
        if j==len( ffList )-1:                                  maxHkls = buffer[ff]['expected']
        buffer[ff]['diff']                                      = np.abs( buffer[ff]['expected'] -  hookLoadArray )      
        if j==0:   bufferArray                                  = buffer[ff]['diff']
        else: bufferArray                                       = np.vstack( ( bufferArray,buffer[ff]['diff'] ) )
    firstKeyFF, lastKeyFF                                       = ( list( buffer.keys(  ) )[0], list( buffer.keys(  ) )[-1] )
    maxDiffHkl                                                  = maxDiffHkl if (buffer[lastKeyFF]['expected'].max( )-buffer[firstKeyFF]['expected'].max( ))<maxDiffHkl else (buffer[lastKeyFF]['expected'].max( )-buffer[firstKeyFF]['expected'].max( ))
    maxDiffHkl                                                  = np.sqrt( maxDiffHkl )
    indices                                                     = np.argmin(  bufferArray, axis=0  )
    calcFFs                                                     = np.array([  float( ffList[i] ) for i in indices  ])
    deltaHkls                                                   = maxHkls-minHkls
    return np.std( calcFFs ),maxDiffHkl,calcFFs,bitDepthsArray,deltaHkls
```

```python
def getAndPlotVariabilityFF( calcFFs,bda,diffs,window=10,outFile="FFVariability.png" ):
    from matplotlib.ticker import (MultipleLocator,AutoMinorLocator)
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sb
    fdict = dict(family="Arial",size=8)
    ser         = pd.Series(calcFFs)
    varis       = ser.rolling( window=window ).std(  )*diffs
    # varis       = varis[varis.notna(  )]
    fig,ax = plt.subplots( figsize=(3.33,2) )
    pal = ["magenta"]
    ax.scatter( x=bda, y=varis, facecolor="magenta", alpha=0.1,edgecolor='black' )
    ax.vlines(bda, ymin=0, ymax=varis, color='magenta',alpha=0.1)
    ax.grid( visible=True,color='lightgrey',ls='dotted', which='both',alpha=0.2 )
    ax.set_ylabel( "Friction \n Variability Factor",fontdict={**fdict, "weight":"bold"} )
    ax.set_xlabel( "Run Measured Depth [ft]",fontdict={**fdict, "weight":"bold"} )
    ax.set_ylim(    0,15000 )
    ax.set_xlim(    0,7600 )
    ax.set_xticklabels( ax.get_xticklabels(  ),fontdict=fdict )
    ax.set_yticklabels( ax.get_yticklabels(  ),fontdict=fdict )
    ax.xaxis.set_minor_locator( AutoMinorLocator( ) )
    ax.yaxis.set_minor_locator( AutoMinorLocator( ) )
    # ax.xaxis.set_label_position( 'top' )
    ax.set_facecolor( 'black' )
    plt.tight_layout( pad=0.3 )
    fig.savefig( outFile,dpi=600 )
```
